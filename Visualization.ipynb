```python
refs = [436,382,826,948,549,346,570,931,353,139]

counter = [229,189,407,466,276,163,296,488,189,79]

stack = counter

aging = [231,193,411,473,284,163,285,480,189,79]

approx = [224,192,406,470,268,160,281,481,188,77]

counter_times = [ 0.008448, 0.007342, 0.015418, 0.018015, 0.010614, 0.005697, 0.009385, 0.013563, 0.005819, 0.002868]

stack_times = [0.006745,0.005958,0.01003,0.016052,0.008371,0.005634,0.009693,0.016763,0.005968,0.00226]

aging_times = [0.008334,0.007787,0.014417,0.018016,0.010313,0.005601,0.009513,0.023142,0.007859,0.002546]

approx_times = [0.007292,0.006563,0.011762,0.013219,0.009143,0.005628,0.009557,0.014177,0.00539,0.002867]
```

```python
counter = collections.OrderedDict(sorted(dict(zip(refs, counter)).items()))
stack = collections.OrderedDict(sorted(dict(zip(refs, stack)).items()))
aging = collections.OrderedDict(sorted(dict(zip(refs, aging)).items()))
approx = collections.OrderedDict(sorted(dict(zip(refs, approx)).items()))

counter_times = collections.OrderedDict(sorted(dict(zip(refs, counter_times)).items()))
stack_times = collections.OrderedDict(sorted(dict(zip(refs, stack_times)).items()))
aging_times = collections.OrderedDict(sorted(dict(zip(refs, aging_times)).items()))
approx_times = collections.OrderedDict(sorted(dict(zip(refs, approx_times)).items()))

```

```python
#Faults
plt.plot([x[0] for x in counter.items()],[x[1] for x in counter.items()],'r',label='Counter')
plt.plot([x[0] for x in stack.items()],[x[1] for x in stack.items()],'g',label='Stack')
plt.plot([x[0] for x in aging.items()],[x[1] for x in aging.items()],'b',label='Aging')
plt.plot([x[0] for x in approx.items()],[x[1] for x in approx.items()],'y',label='Approx')
plt.xlabel('No. of references', fontsize=20)
plt.ylabel('No. of faults', fontsize=20)
plt.legend(prop={'size':25})
plt.rcParams['figure.figsize'] = [20,10]
plt.savefig('fault_graph.png')
plt.show()
```
